## Grouping in Pandas

#### Single Column Grouping with single aggregation

```python
# copying first 10 records from df2
y= df2.copy().head(10)

# adding new column budget with data
y['budget']=[1000,2000,1200,7000,3232,909,700,8909,10000,9090]

# adding new column boxoffice collection
y['boxoffice']=[2000,3000,700,800,200,1000,700,800,200,1000]

# displaying data frame
print(y)
```

```python
# Finding how many records each unique year contain for total values
print("Total Values")
print(y.groupby('year').count())
```

```python
# Finding how many records each unique year contain for first 5 values
print("First 5 Values")
print(y.groupby('year').count().head())
```

```python
# Finding how many records each unique year contain - it dont show for each column
y.groupby('year').size()
```

```python
# other method of grouping - like groupby('month')['income'].mean()
y.groupby('title')['year'].mean()
```

#### Multi Column Grouping with single aggregation

```python
# grouping multiple columns
y.groupby(['title','year'])['budget'].sum()
```

```python
# checking count of unique titles - since all are unique, sum giving same value above
y.title.nunique()
```

```python
# method 2 grouping multiple columns
y.groupby(['title','year']).budget.sum()
```

#### multiple aggregations

```python
# single column with multiple aggrigations - like pivote table can be seen
y.groupby('year').agg({'budget':['sum','count','min','max'],'boxoffice':['std','var','mean','median']})
```

```python
# multiple columns with multiple aggregations
y.groupby(['title','year']).agg({'budget':'sum','boxoffice':'mean'})
```

```python
# method 2 using funtions multiple column grouping with multiple aggregations
y.groupby(['title','year']).agg({'budget':sum,'boxoffice':[max,min]})
```

#### Using Numpy Functions

```python
# using numpy functions - as if we used functions above only sum,max,min available not var, std, mean etc
#import numpy
import numpy as np

#using grouping
y.groupby('year').agg({'budget':np.mean,'boxoffice':[np.median,np.var,np.std]})
```

## Bins in Pandas and Grouping Ranges

#### Using pd.cut() of Pandas

```python
# Taking 50 Values from Original Data Frame to Demonstrate
bins_df=df2.head(50).copy()
print(bins_df)
```

```python
# To create bins check max and min values for which column required to be in bin range
bins_df.describe() # found 1916 to 2017 values available, so create bins from 1916 to 2017 only is enough
```

```python
# Creating bins
bins = [1910,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020,2050]
bins_df['year_range']=pd.cut(bins_df['year'],bins)
print(bins_df)
```

```python
# creating bins with labels
bins2 = [1910,1940,1970,2000,2030]
bins_df['range_label']=pd.cut(bins_df['year'],bins2,labels=['Early < 1940','mid < 1970','high < 2000','above < 2030'])
bins_df
```

```python
# Checking how many records falls in different bin labels
bins_df.range_label.value_counts()
```

#### Using Python Function

```python
#creating function
def bins3(y):
    if y>=1910 and y<=1950:
        return "Early Years"
    elif y>=1951 and y<=2000:
        return "Mid Years"
    else:
        return "21st Century year"

# cross checking by giving 1 test values
print("Test value given 2000, and year range assigned: ",bins3(2000))

# Applying for all the series values for year after success of above test
bins_df['century years']=bins_df['year'].apply(lambda x:bins3(x))
```

```python
# Displaying new bins_df dataframe
bins_df
```

```python
# Checking how many records falls in different century years label
bins_df['century years'].value_counts()
```

## Joins

```python
# can join using merge, join etc
# creating new data frame 1
df1=pd.DataFrame({'Roll_id':[1,2,3,4,5],'age':[10,20,30,40,50]})
print(df1)

print("\n")

# creating new data frame 2
df2=pd.DataFrame({'id':[6,7,1,2,3],'income':[100,200,300,400,500]})
print(df2)
```

#### inner join

```python
# common value from both tables
# METHOD 1
df1.merge(df2,how='inner',left_on='Roll_id',right_on='id')
```

```python
# METHOD 2
pd.merge(df1,df2,how='inner',left_on='Roll_id',right_on='id')
```

#### left join

```python
# all values from left table and common values in 2nd table with first table
# METHOD 1
df1.merge(df2,how='left',left_on='Roll_id',right_on='id')
```

```python
# METHOD 2
pd.merge(df1,df2,how='left',left_on='Roll_id',right_on='id')
```

#### right join

```python
# all values from right table and common values in 1st table with second table
# METHOD 1
df1.merge(df2,how='right',left_on="Roll_id",right_on='id')
```

```python
# METHOD 2
pd.merge(df1,df2,how='right',left_on='Roll_id',right_on='id')
```

#### outer join

```python
# all the values from all the tables - just like union in set
# METHOD 1
df1.merge(df2,how='outer',left_on='Roll_id',right_on='id')
```

```python
# METHOD 2
pd.merge(df1,df2,how='outer',left_on='Roll_id',right_on='id')
```

#### cross join

```python
# each column in table 1 with each column in table 2
# usefull: 1. month in one table, days in one table and get transactions of each month comparing each days in second table
#          2. days in one table, days in other table, get compare each day in table 1 with each day in table 2 sales 
# create keys for this in both tables, after join drop the column keys

df1['key']=1
df2['key']=1
print(df1)

print("\n")

print(df2)
```

```python
# cross joining 
# METHOD 1 - auto suffix came as _x and _y, drop if dont required
df1.merge(df2,how='cross')
```

```python
# METHOD 2
pd.merge(df1,df2,how='cross')
```

```python
# METHOD 3 - if both column key name is same give just on, default join is inner join , here work like cross join
# remove key column if not required
pd.merge(df1,df2,on='key')
```
